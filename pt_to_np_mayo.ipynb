```python
import torch
import matplotlib.pyplot as plt 
import numpy as np
import os
import cv2
from torchvision import transforms 
#resize = transforms.Resize(1024)
```

```python
def map_to_zero_one_alt(X, min_v = None, max_v = None, return_vals=False):
  if min_v == None:
      min_val = X.min()
      max_val = X.max()
  else:
      min_val = min_v
      max_val = max_v
  X = (X-min_val)/(max_val-min_val)
  if return_vals:
      return X, min_val, max_val
  else:
      return X
```

```python
def inv_map_to_zero_one_alt(X, min_val, max_val):
  """Inverse data normalizer (from [0,1])"""
  return X*(max_val-min_val)+min_val
```

```python
class_name = 'HD'
suff = 'train_mayo_3_alt'

temp = torch.load('./datasets_unzipped/'+suff+'.pt')
print(temp.size())

save_path = './datasets_unzipped/'+suff+'/'+class_name+'/'
try:
    os.makedirs(save_path)
except FileExistsError:
    pass
min_val = torch.tensor(-1024.)
max_val = torch.tensor(3072.)
temp = map_to_zero_one_alt(temp[:,0:1,:,:],min_val,max_val)
print(temp.size(0))
for i in range(temp.size(0)):
    img = temp[i,0:1,:,:]
    img = torch.permute(img,(1,2,0))
    print(img.min())
    print(img.max())
    img = img.numpy()
    print(img.shape)
    np.save(save_path+ f'{i+1:08d}'+'.npy',img)
minmax = torch.cat((min_val.unsqueeze(0),max_val.unsqueeze(0)),0)
torch.save(minmax,'./datasets_unzipped/'+suff+'_minmax.pt')
```
